```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import os

def gauss_fit(x,mu,sigma,C,A):
    y = C+(A*(np.exp((-1*(x-mu)**2)/(2*sigma**2))))
    return y

sigma_row_AllFiles = []
sigma_column_AllFiles = []

import glob
import matplotlib.pyplot as plt
os.chdir(r"C:\QuMix\Anirban\2024-06-26\dataset_1")
directory = glob.glob('*.csv')

for files in directory:
    filename = files
    os.chdir(r"C:\QuMix\Anirban\2024-06-26\dataset_1")
    df = pd.read_csv(filename)
    name_string = files[5:9]
    row_name = name_string + '_row'
    column_name = name_string + '_col'
    row_sums = df.sum(axis=1).tolist()
    column_sums = df.sum(axis=0).tolist()
    
    # matrix_data = np.genfromtxt(filename, delimiter=',')  # Replace 'matrix.csv' with your CSV file path

    # #Create a heatmap plot
    # plt.figure(figsize=(8, 6))  # Adjust size as needed
    # plt.imshow(matrix_data, cmap='viridis', interpolation='nearest')
    # plt.colorbar(label='Color Scale')  # Add color bar indicating values

    # plt.title(name_string)  # Set plot title
    # # plt.xlabel('X Axis')  # Set X axis label
    # # plt.ylabel('Y Axis')  # Set Y axis label
    # plt.grid(False)  # Disable grid for heatmap

    # # Step 3: Save the heatmap plot as an image file
    # os.chdir(r'C:\Users\srishic\Desktop\2024-06-26\Figures')
    # plt.savefig(name_string + '.jpg')  # Save as JPG image (or choose another format)
    # #plt.show()
    # #plt.close()  # Close the plot to free memory

    # print("Sum of each row:", row_sums)
    # print("Sum of each column:", column_sums)
    row_sum_partial = row_sums[800:1300]
    #row_sum_partial = (row_sum_norm-np.min(row_sum_norm))
    #row_sum_partial = row_sum_partial/np.max(row_sum_partial)
    x = np.arange(len(row_sum_partial))
    #y = gauss_fit(x,1,1,0.2,0)


    popt, pcov = curve_fit(gauss_fit, x, row_sum_partial,p0=[250,100,2.2e6,0],maxfev=100000)
    y_fit = gauss_fit(x,popt[0],popt[1], popt[2], popt[3])
    y_fit_norm = y_fit/np.max(y_fit)
    #fig = plt.figure()
    #plt.scatter(x,row_sum_partial)
    #plt.plot(x, y_fit_norm,'r-')
    # os.chdir(r'C:\Users\srishic\Desktop\2024-06-26\Figures')
    # plt.savefig( row_name +'.jpg')

    # imshow(df)
    sigma_row_AllFiles.append(np.abs(popt[1]))
    print(f'The sigma for the fitting is {np.abs(popt[1])}')

    column_sum_partial = column_sums[500:1000]
    #column_sum_partial = (column_sum_norm-np.min(column_sum_norm))
    #column_sum_partial = column_sum_partial/np.max(column_sum_partial)
    x = np.arange(len(column_sum_partial))


    popt, pcov = curve_fit(gauss_fit, x, column_sum_partial,p0=[100,20,2e6,0],maxfev=100000)
    y_fit = gauss_fit(x,popt[0],popt[1],popt[2], popt[3])
    y_fit_norm = y_fit/np.max(y_fit)
    #fig = plt.figure()
    #plt.scatter(x,column_sum_partial)
    #plt.plot(x, y_fit_norm,'r-')
    # os.chdir(r'C:\Users\srishic\Desktop\2024-06-26\Figures')
    # plt.savefig( column_name +'.jpg')

    sigma_column_AllFiles.append(np.abs(popt[1]))
    print(f'The sigma for the fitting is {np.abs(popt[1])}')


#sigma_column_AllFiles
#sigma_row_AllFiles
```
